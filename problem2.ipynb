**3** Χρησιμοποιώντας τα δεδομένα εκπαίδευσης, εκπαιδεύστε δύο δένδρα παλινδρόμησης με κριτήρια διαχωρισμού τα squared_eror και poisson ([δείτε ενότητα 1.10.7.2](https://scikit-learn.org/stable/modules/tree.html#tree)) αντίστοιχα, θέτοντας τον σπόρο τυχαιότητας στην τιμή 42 και το μέγιστο βάθος στην τιμή 3. Υπολογίστε τη μετρική r2 με βάση τις προβλέψεις τους στα δεδομένα ελέγχου, τις οποίες αποθηκεύστε στις μεταβλητές y_pred1 (squared_error) και y_pred2 (poisson) αντίστοιχα.
Αποθηκεύστε τις μετρικές στις μεταβλητές r2_squared_error και r2_poisson αντίστοιχα. (1 μονάδα)

```python
# YOUR CODE HERE
# ---------------

from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

max_depth = 3
# Init. Decision Tree Regressor with squared_error criterion
regressor_model_squared_error = DecisionTreeRegressor(criterion='squared_error',
                                                      random_state=42,
                                                      max_depth=max_depth)
# Init. Decision Tree Regressor with poisson criterion
regressor_model_poisson = DecisionTreeRegressor(criterion='poisson',
                                                random_state=42,
                                                max_depth=max_depth)

# Fit the regressor to the training data on both models
regressor_model_squared_error.fit(X_train, y_train)
regressor_model_poisson.fit(X_train, y_train)

# Predict the target values for the test data using both models
y_pred1 = regressor_model_squared_error.predict(X_test)
y_pred2 = regressor_model_poisson.predict(X_test)

# Calculate the R^2 score of both models
r2_squared_error = r2_score(y_test, y_pred1)
r2_poisson = r2_score(y_test, y_pred2)
```

Σε περίπτωση που ελέγχεται δεύτερη φορά μια μεταβλητή στο μονοπάτι, θα πρέπει να διατηρούμε τον τελευταίο και στενότερο περιορισμό. Ακολουθούν παραδείγματα ορθής λειτουργίας, στα οποία φαίνεται και η μορφή αν-τότε που πρέπει να έχει το μονοπάτι. (4 μονάδες)

```python
import numpy as np

def get_path(tree, instance, feature_names):
    # YOUR CODE HERE
    # ---------------

    node_indicator = tree.decision_path([instance]) # Get the decision path of the instance with all the nodes it visited (sparse matrix)
    feature = tree.tree_.feature                    # Get the feature names of the tree
    threshold = tree.tree_.threshold                # Get the threshold values of the tree

    # Find the node index for the given instance on the top
    node_index = node_indicator.indices[node_indicator.indptr[0]:node_indicator.indptr[1]][0]

    path = []   # List to hold the substrings of the path rules
    # Iterate from the top of the tree to the bottom visiting each node (it did visit) in the path
    while True:
        # Find the threshold sign for the current node
        # If the feature value of the instance is less than or equal to the threshold, the sign is "<="
        # Otherwise, the sign is ">"
        if instance[feature[node_index]] <= threshold[node_index]:
            threshold_sign = "<="
        else:
            threshold_sign = ">"

        # If the current node is a leaf node, break the loop since we don't want to print the threshold for the leaf node
        # since it doesn't have a threshold and it doesn't get checked
        if tree.tree_.children_left[node_index] == tree.tree_.children_right[node_index]:
            break

        # Check if there are similar rules (with the same feature and sign) already in the path string
        # If there are rules, then we are about to add is a "stricker" version of the last and need to delete the older version
        # For that we iterate over all the rules in the path list and check if there is a rule for the same feature with the same sign
        # If we find that we have added a rule for the same feature with the same sign, we pop the last rule from the path list to insert the current stricker rule
        for i in range(len(path)):
            if f"{feature_names[feature[node_index]]} {threshold_sign}" in path[i]:
                # Pop the last rule from the path list since the current rule is FOR SURE a "stricker" version of the last
                # since we go from the top of the tree to the bottom
                path.pop(i)
                # By observing the expected output, they don't seem to be sort by feature name, so we don't need to do anything further
        # Finally, we append the latest rule to the path list
        path.append(f"{feature_names[feature[node_index]]} {threshold_sign} {round(threshold[node_index], 2)}")
        # {threshold[node_index]:.2f}

        # Find the next node index based on the threshold value of the current node
        # If the feature value of the instance is less than or equal to the threshold, the next node index is the left child
        if instance[feature[node_index]] <= threshold[node_index]:
            # Update the node index to the next node index
            node_index = tree.tree_.children_left[node_index]

        # Otherwise, the next node index is the right child
        else:
            # Update the node index to the next node index
            node_index = tree.tree_.children_right[node_index]

    # Join the rules with the word "and" (in Greek)
    rules_path_str = " και ".join(path)

    # Calculate the prediction of the instance using the tree
    # and round it to 2 decimal points to pass the tests
    prediction = tree.predict([instance])[0]
    prediction = round(prediction, 2)

    # Return the rules string by adding the "if" and "then" parts (in Greek) in the front and the end of the string accordingly
    # Also we append the prediction of the instance
    return "αν " + rules_path_str + " τότε " + str(prediction)
```

**6** Υλοποιήστε συνάρτηση, η οποία δοθέντος ενός δενδρικού μοντέλου παλινδρόμησης, διερευνά όλες τις ακέραιες τιμές της μεταβλητής min_samples_leaf από ένα κάτω όριο μέχρι ένα άνω όριο και επιστρέφει την βέλτιστη με βάση το mean squared error στο παραπάνω σύνολο επικύρωσης έπειτα από εκπαίδευση στο παραπάνω σύνολο εκπαίδευσης. Επιπλέον επιστρέφει και το mean squared error (στρογγυλοποιημένο σε ακέραιο) που πετυχαίνει στο σύνολο ελέγχου το μοντέλο με την βέλτιστη επίδοση στο σύνολο επικύρωσης. (2 μονάδες)

```python
from sklearn.metrics import mean_squared_error

def best_min_samples_leaf(model, min, max):
    # YOUR CODE HERE
    # ---------------

    # Initialize the best mse with infinity value to ensure that the first mse will be less than it
    best_mse = float('inf')
    # Initialize the best leaf with None value to ensure that the first leaf will be the best (temporary)
    best_leaf = None

    # Iterate over all the possible values of min_samples_leaf
    # and find the best one that minimizes the mse on the validation set
    for leaf in range(min, max+1):
        # Set the current min_samples_leaf to the leaf value in the current iteration
        model.min_samples_leaf = leaf

        # Fit the model on the training set with the current leaf value
        model.fit(X_train, y_train)

        # Predict the validation set with the current leaf value
        y_pred_val = model.predict(X_val)

        # Calculate the mse on the validation set on the current leaf value
        mse_val = mean_squared_error(y_val, y_pred_val)

        # Update the best mse and the best leaf if the current mse is less than the best mse
        if mse_val < best_mse:
            best_mse = mse_val
            best_leaf = leaf

    # Since we found the best leaf, we set it to the model
    # and refit the model on the training set
    model.min_samples_leaf = best_leaf
    model.fit(X_train, y_train)

    # Predict the test set with the best model (the one with the best leaf value)
    y_pred_test = model.predict(X_test)

    # Calculate and round the test mse into an integer
    mse_test = mean_squared_error(y_test, y_pred_test)
    mse_test = round(mse_test)

    # Return the best leaf and the test mse
    return best_leaf, mse_test
```

```python
"""Τεστ ορθής υλοποίησης συνάρτησης"""
model = DecisionTreeRegressor(max_depth=3, random_state=42)
assert best_min_samples_leaf(model, 1, 8) == (7, 3358)
assert best_min_samples_leaf(model, 5, 15) == (11, 3271)
```

**7** Υλοποιήστε συνάρτηση η οποία δοθέντος ενός μοντέλου παλινδρόμησης, ενός συνόλου δεδομένων και μίας μετρικής, το αξιολογεί με τη μέθοδο της σταυρωτής επικύρωσης "άφησε ένα εκτός". (1 μονάδα)

```python
from sklearn.model_selection import LeaveOneOut
import numpy as np

def leave_one_out(model, X, y, metric):
    # YOUR CODE HERE
    # ---------------

    # Initialize the Leave-One-Out cross-validator object
    # Provides train/test indices to split data in train/test sets.
    # Each sample is used once as a test set (singleton) while the remaining samples form the training set.
    loo_obj = LeaveOneOut()

    scores = []     # Initialize an empty list to store the scores of the model
    # Iterate over all the possible train/test indices of the Leave-One-Out cross-validator object
    # and keep the indices of the training and test set
    for train_index, test_index in loo_obj.split(X):
        # Save temporarily the training and test set
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Fit the model on the temp. training set from above
        model.fit(X_train, y_train)
        # Predict the test set with the model
        y_pred = model.predict(X_test)

        # Calculate the score of the model on the test set and append it to the scores list
        score = metric(y_test, y_pred)
        scores.append(score)

    # Calculate the mean of the scores list and return it
    mean_score = np.mean(scores)
    return mean_score
```
